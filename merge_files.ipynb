# Tableau Homework CitiBike Analysis

Data Retrieved From: https://www.citibikenyc.com/system-data - https://s3.amazonaws.com/tripdata/index.html
Data Format: Indiviual zipped csv files for each month JAN-DEC of year 2019

This notebook is used to combine this data into one large dataset that can easily be used in Tableu without the need to join each monthly file individually. This notebook is also used to reduce the size of the final dataset as Tableu public limits the data size of each workbook to 15,000,000 rows and this dataset is in excess of 20,000,000. 

To reduce the size of the dataset any rows with missing gender data were removed. The birth year column was removed as I did not use that in my analysis. I also removed rows with trip duration of less than 6 minutes to get the dataset down to the appropriate size. I also believe it is unlikely many users take rides of less than 6 minutes so some of this may be suspicious data that I felt was save to remove.

```python
#found a way to merge many files at once with something called glob

import pandas as pd
import glob
```

```python
citi_files = glob.glob("*.csv") 
df = pd.concat((pd.read_csv(f, header = 0) for f in citi_files))
df
```

```python
# Table is too big for Tableau public so deleting a column I dont need
df = df.drop(columns=['birth year', 'Unnamed: 0'])
df
```

```python
# drop rows with no gender data to decrease dataset size
index_names = df[(df['gender']< 1)].index 
df.drop(index_names, inplace = True) 
df 
```

```python
# drop rows with low trip duration <6 minutes to decrease dataset size
index_names = df[(df['tripduration']< 360)].index 
df.drop(index_names, inplace = True) 
df 
```

```python
# Export file to csv
df.to_csv("2019-citibikedata_combined.csv")
```
